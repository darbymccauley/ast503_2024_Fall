````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.integrate import simps
from astropy.cosmology import FlatLambdaCDM

# Define cosmology
cosmo = FlatLambdaCDM(H0=70, Om0=0.3)

class Spectrum:
    def __init__(self, wavelength=None, frequency=None, flux=None, flux_unit='W/m^2/m'):
        if wavelength is not None:
            self.wavelength = np.array(wavelength)
            self.frequency = 299792458 / self.wavelength  # c = 299792458 m/s
        elif frequency is not None:
            self.frequency = np.array(frequency)
            self.wavelength = 299792458 / self.frequency
        else:
            raise ValueError("Either wavelength or frequency must be provided.")

        self.flux = np.array(flux)
        self.flux_unit = flux_unit

        # Interpolation functions
        self._flux_interp_wavelength = interp1d(
            self.wavelength, self.flux, kind='linear', bounds_error=False, fill_value='extrapolate'
        )
        self._flux_interp_frequency = interp1d(
            self.frequency, self.flux, kind='linear', bounds_error=False, fill_value='extrapolate'
        )

    def flux_at_wavelength(self, wavelength):
        return self._flux_interp_wavelength(wavelength)

    def flux_at_frequency(self, frequency):
        return self._flux_interp_frequency(frequency)

class Filter:
    def __init__(self, wavelength=None, frequency=None, response=None):
        if wavelength is not None:
            self.wavelength = np.array(wavelength)
            self.frequency = 299792458 / self.wavelength
        elif frequency is not None:
            self.frequency = np.array(frequency)
            self.wavelength = 299792458 / self.frequency
        else:
            raise ValueError("Either wavelength or frequency must be provided.")

        self.response = np.array(response)

        # Interpolation functions
        self._response_interp_wavelength = interp1d(
            self.wavelength, self.response, kind='linear', bounds_error=False, fill_value=0.0
        )
        self._response_interp_frequency = interp1d(
            self.frequency, self.response, kind='linear', bounds_error=False, fill_value=0.0
        )

    def response_at_wavelength(self, wavelength):
        return self._response_interp_wavelength(wavelength)

    def response_at_frequency(self, frequency):
        return self._response_interp_frequency(frequency)

class StandardSource(Spectrum):
    pass

class KCorrectionCalculator:
    def __init__(self, observed_spectrum, redshift, filter_R, standard_source_R, use_wavelength=True):
        self.observed_spectrum = observed_spectrum
        self.z = redshift
        self.filter_R = filter_R
        self.standard_source_R = standard_source_R
        self.use_wavelength = use_wavelength

    def compute_k_correction_for_band(self, filter_Q, standard_source_Q):
        self.filter_Q = filter_Q
        self.standard_source_Q = standard_source_Q
        K_QR = self.compute_k_correction()
        return K_QR

    def compute_k_correction(self):
        if self.use_wavelength:
            lambda_obs = self.filter_R.wavelength
            lambda_emit = lambda_obs / (1 + self.z)

            f_lambda_obs = self.observed_spectrum.flux_at_wavelength(lambda_obs)
            f_lambda_emit = self.observed_spectrum.flux_at_wavelength((1 + self.z) * lambda_emit)
            R_lambda = self.filter_R.response
            Q_lambda = self.filter_Q.response_at_wavelength(lambda_emit)
            g_R_lambda = self.standard_source_R.flux_at_wavelength(lambda_obs)
            g_Q_lambda = self.standard_source_Q.flux_at_wavelength(lambda_emit)

            # Use simps for integration
            N1 = simps(lambda_obs * f_lambda_obs * R_lambda, lambda_obs)
            N2 = simps(lambda_emit * g_Q_lambda * Q_lambda, lambda_emit)
            D1 = simps(lambda_obs * g_R_lambda * R_lambda, lambda_obs)
            D2 = simps(lambda_emit * f_lambda_emit * Q_lambda, lambda_emit)

            K_factor = (1 + self.z) * (D1 * D2) / (N1 * N2)
        else:
            pass  # Frequency domain calculations if needed

        # Compute K-correction and ensure it is positive
        K_QR = -2.5 * np.log10(K_factor)
        return K_QR

# Load LSST filters
bands = ['u', 'g', 'r', 'i', 'z', 'y']
filters = {}
for band in bands:
    file = pd.read_csv(f'./data/LSST_LSST.{band}_filter.dat', delimiter=' ')
    filt_data = np.array(file.values)
    wavelength_angstrom = filt_data[:, 0]  # Wavelength in Angstroms
    response = filt_data[:, 1]
    wavelength_meters = wavelength_angstrom * 1e-10  # Convert from Angstroms to meters
    filters[band] = Filter(wavelength=wavelength_meters, response=response)

# Effective wavelengths in meters
lam_eff_microns = [0.3671, 0.4827, 0.6223, 0.7546, 0.8691, 0.9712]  # in micrometers
lam_eff = np.array(lam_eff_microns) * 1e-6  # Convert to meters
eff_wavelengths = dict(zip(bands, lam_eff))

# Load observed SED
data = np.loadtxt('aless_average_seds.dat')[:, :2]
observed_wavelength_microns = data[:, 0]  # Wavelength in micrometers
observed_wavelength = observed_wavelength_microns * 1e-6  # Convert to meters
observed_flux_nu_mJy = data[:, 1]  # F_nu in mJy

# Convert F_nu from mJy to W/m^2/Hz
observed_flux_nu = observed_flux_nu_mJy * 1e-26  # Convert mJy to W/m^2/Hz

# Convert F_nu to F_lambda
c = 299792458  # Speed of light in m/s
observed_flux_lambda = (observed_flux_nu * c) / (observed_wavelength ** 2)  # W/m^2/m

# Create sed array
sed = np.column_stack((observed_wavelength, observed_flux_lambda))

# Redden the SED using your provided function
def redden_SED(sed, redshift):
    sed_z = np.zeros(sed.shape)
    sed_z[:,0] = sed[:,0] * (1 + redshift)  # Wavelength, multiply by (1+z)
    D_L = cosmo.luminosity_distance(redshift).value * 3.0857e22  # D_L in meters
    sed_z[:,1] = sed[:,1] * (1e-5 / D_L)**2 * (1 + redshift)
    return sed_z

# Define the AB standard source
def AB_func(wavelength):
    F_nu = 3631e-26  # Flux in W/m^2/Hz (3631 Jy)
    c = 299792458    # Speed of light in m/s
    F_ab_flux_lambda = (F_nu * c) / (wavelength ** 2)
    return Spectrum(wavelength=wavelength, flux=F_ab_flux_lambda)

standard_wavelength = np.linspace(300e-9, 1100e-9, 5000)  # Wavelengths from 300 nm to 1100 nm
standard_source = AB_func(standard_wavelength)

# Prepare redshift array
z_array = np.arange(0, 2.01, 0.01)
observer_bands = ['g', 'i', 'y']
K_corrections = {band: [] for band in observer_bands}
z_valid = {band: [] for band in observer_bands}

# Main computation loop
for z in z_array:
    # Redshift the observed SED
    sed_z = redden_SED(sed, z)
    sed_z_wavelength = sed_z[:,0]
    sed_z_flux = sed_z[:,1]
    observed_spectrum = Spectrum(wavelength=sed_z_wavelength, flux=sed_z_flux)

    for obs_band in observer_bands:
        lambda_eff_obs = eff_wavelengths[obs_band]  # Observer-frame effective wavelength
        lambda_eff_rest = lambda_eff_obs / (1 + z)  # Rest-frame effective wavelength

        # Find the rest-frame band with closest effective wavelength
        rest_bands = ['u', 'g', 'r', 'i', 'z', 'y']
        rest_lambdas = np.array([eff_wavelengths[band] for band in rest_bands])
        delta_lambda = np.abs(rest_lambdas - lambda_eff_rest)
        min_index = np.argmin(delta_lambda)
        min_band = rest_bands[min_index]

        # Check if mapping is valid
        if lambda_eff_rest < rest_lambdas.min() or lambda_eff_rest > rest_lambdas.max():
            continue  # Stop computing for this band at this redshift
        else:
            filter_R = filters[obs_band]
            standard_source_R = standard_source
            filter_Q = filters[min_band]
            standard_source_Q = standard_source

            k_correction_calculator = KCorrectionCalculator(
                observed_spectrum=observed_spectrum,
                redshift=z,
                filter_R=filter_R,
                standard_source_R=standard_source_R,
                use_wavelength=True
            )

            K_QR = k_correction_calculator.compute_k_correction_for_band(filter_Q, standard_source_Q)
            if K_QR >= 0:
                K_corrections[obs_band].append(K_QR)
                z_valid[obs_band].append(z)
            else:
                continue  # Skip negative K-corrections

# Plotting the K-correction curves
plt.figure(figsize=(10, 6))
colors = {'g': 'green', 'i': 'red', 'y': 'blue'}
for band in observer_bands:
    plt.plot(z_valid[band], K_corrections[band], label=f'{band}-band', color=colors[band])
plt.xlabel('Redshift z')
plt.ylabel('K-correction (mag)')
plt.title('K-correction vs Redshift for LSST g, i, y bands')
plt.legend()
plt.grid(True)
plt.show()
````

````python
import numpy as np
from scipy.interpolate import interp1d

def interpolate(x_known, y_known, x_new, kind='linear', fill_value='extrapolate'):
    interpolation_function = interp1d(x_known, y_known, kind=kind, fill_value=fill_value)
    y_new = interpolation_function(x_new)
    return y_new

def integrate(x, y):
    return np.trapz(y, x)

def compute_k_correction(f_obs, z, R, Q, g_R, g_Q, use_wavelength=False):
    """
    Compute the K-correction K_{QR}.

    Parameters:
    - f_obs: Observed spectrum object with attributes wavelength/frequency and flux.
    - z: Redshift of the source.
    - R: Filter R object with attributes wavelength/frequency and response.
    - Q: Filter Q object with attributes wavelength/frequency and response.
    - g_R: Standard source flux for band R.
    - g_Q: Standard source flux for band Q.
    - use_wavelength: Boolean flag to use wavelength domain.

    Returns:
    - K_QR: The K-correction value.
    """
    if use_wavelength:
        # Wavelength domain calculations
        lambda_obs = R.wavelength  # Wavelength grid for filter R
        lambda_emit = lambda_obs / (1 + z)  # Emitted wavelengths

        # Interpolate observed flux onto lambda_obs
        f_lambda_obs = interpolate(f_obs.wavelength, f_obs.flux, lambda_obs)
        # Interpolate observed flux onto (1 + z) * lambda_emit
        f_lambda_emit = interpolate(f_obs.wavelength, f_obs.flux, (1 + z) * lambda_emit)
        # Filter responses (already on lambda_obs and lambda_emit)
        R_lambda = R.response
        Q_lambda = interpolate(Q.wavelength, Q.response, lambda_emit)
        # Standard source fluxes
        g_R_lambda = interpolate(g_R.wavelength, g_R.flux, lambda_obs)
        g_Q_lambda = interpolate(g_Q.wavelength, g_Q.flux, lambda_emit)

        # Compute Numerators and Denominators
        N1 = integrate(lambda_obs, lambda_obs * f_lambda_obs * R_lambda)
        N2 = integrate(lambda_emit, lambda_emit * g_Q_lambda * Q_lambda)
        D1 = integrate(lambda_obs, lambda_obs * g_R_lambda * R_lambda)
        D2 = integrate(lambda_emit, lambda_emit * f_lambda_emit * Q_lambda)

        # Compute K-factor and K-correction
        K_factor = (1 / (1 + z)) * (N1 * N2) / (D1 * D2)
    else:
        # Frequency domain calculations
        nu_obs = R.frequency  # Frequency grid for filter R
        nu_emit = (1 + z) * nu_obs  # Emitted frequencies

        # Interpolate observed flux onto nu_obs
        f_nu_obs = interpolate(f_obs.frequency, f_obs.flux, nu_obs)
        # Interpolate observed flux onto nu_emit / (1 + z)
        f_nu_emit = interpolate(f_obs.frequency, f_obs.flux, nu_emit / (1 + z))
        # Filter responses (already on nu_obs and nu_emit)
        R_nu = R.response
        Q_nu = interpolate(Q.frequency, Q.response, nu_emit)
        # Standard source fluxes
        g_R_nu = interpolate(g_R.frequency, g_R.flux, nu_obs)
        g_Q_nu = interpolate(g_Q.frequency, g_Q.flux, nu_emit)

        # Compute Numerators and Denominators
        N1 = integrate(nu_obs, (1 / nu_obs) * f_nu_obs * R_nu)
        N2 = integrate(nu_emit, (1 / nu_emit) * g_Q_nu * Q_nu)
        D1 = integrate(nu_obs, (1 / nu_obs) * g_R_nu * R_nu)
        D2 = integrate(nu_emit, (1 / nu_emit) * f_nu_emit * Q_nu)

        # Compute K-factor and K-correction
        K_factor = (1 + z) * (N1 * N2) / (D1 * D2)

    # Compute K-correction
    K_QR = -2.5 * np.log10(K_factor)
    return K_QR
````
